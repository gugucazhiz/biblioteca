```python
from pathlib import Path
import pandas as pd
```

```python
df = pd.read_csv(
    Path.cwd() / 'bd/biblioteca.csv',
    sep=",",
    encoding="latin1"
)


#df.to_excel("biblioteca.xlsx",index=False,engine="openpyxl")
df.head()
```

```python
# transformando csv em xlsx
df2 = pd.read_excel(
    Path.cwd() / 'bd/biblioteca.xlsx',
    engine='openpyxl',
)
#FIltrando faculdade macaiba
macaiba = df2[(df2['biblioteca'] == "Biblioteca Setorial Prof. Rodolfo Helinski - Escola AgrÃ­cola de JundiaÃ­ - EAJ - Macaiba")]

#macaiba.to_excel("macaiba.xlsx",index=False,engine="openpyxl")
macaiba.head()
```

```python
# percebi que nesse dataset tem valores duplicados.
# Esses valores duplicados são propositais, para cada livro que ira ser cadastrado
# mesmo que ja exista um identico dele no sistema, a copia nova sera adicionada ao sistema usando
# o mesmo "id" do original/primeiro.

# removendo os valores duplicados

macaiba_sem_dup = macaiba.drop_duplicates(subset="registro_sistema",keep="first")
macaiba_sem_dup.head()

#macaiba_sem_dup.to_excel("macaiba1.xlsx",index=False,engine="openpyxl")
```

```python
livros = pd.read_csv(
    Path.cwd() / "bd/livros.csv",
    encoding='latin1',
    sep=";",
    engine= "python"
)
#precisei usar o engine python no csv por causa de linhas quebradas ou caracteres especiais
livros.head(10)
#df.to_excel("livros.xlsx",index=False,engine="openpyxl")
```

```python
#registro_sistema

juncao = livros.merge(macaiba_sem_dup,on="registro_sistema",how="inner")

juncao.head()

juncao.to_excel("juncao.xlsx",index=False,engine="openpyxl")
```
